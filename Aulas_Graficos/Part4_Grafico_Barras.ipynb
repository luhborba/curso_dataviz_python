```python
# Imports
import plotly.express as px
```

```python
#Criando DataSet do Plotly
data_br = px.data.gapminder().query('country == "Brazil"')
data_br.head()
```

```python
# Criando Gráfico de Forma Simples
fig1 = px.bar(data_br, x='year', y='pop')
fig1.show()
```

```python
# Personalizando o Gráfico
fig1 = px.bar(data_br, x='year', y='pop',
              title = 'População Brasileira')

fig1.update_layout(
    xaxis = dict(title='Anos'),
    yaxis = dict(title='População'),
)

fig1.show()
```

```python
#Criando Gráfico sobre expectativa de vida
fig2 = px.bar(data_br, x='year', y='lifeExp',
              title = 'Expectativa de vida da população brasileira')

fig2.update_layout(
    xaxis = dict(title='Anos'),
    yaxis = dict(title='Expectativa de Vida'),
)

fig2.show()
```

```python
# Criando grafico de Barras Horizontais, que mostra a expctativa de vida, também trás a informação que quando a expectativa aumento, influênciou no aumento da população.
fig3 = px.bar(data_br, x='lifeExp', y='year', orientation='h',
              labels = {'lifeExp':'Expectativa de Vida',
                        'year':'Ano'},
              title = 'Expectativa de Vida no Brasil (1952-2007)',
              height = 400,
              color = 'pop'
              )
fig3.show()
```
