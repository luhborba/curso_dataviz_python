## Etapa 9: Gráficos Estatísticos

```python
# Importações
import plotly.express as px
import pandas as pd
import numpy as np
```

```python
# Criando conjunto de dados aleatórios com 'numpy' e transformando em DataFrame
data = {
    'A': np.random.normal(10, 0.5, 10_000),
    'B': np.random.normal(10, 2.5, 10_000)
}

df = pd.DataFrame(data)
df.head()
```

```python
# Criando Histograma de Frequência
fig = px.histogram(df, x='A', title = 'Histograma da Variável A',
                   nbins = 20) # Está função define a quantidade de grupos para o gráfico
fig.update_layout(
    xaxis = dict(title = 'Variável A'),
    yaxis = dict(title = 'Frequência'),
)
fig.show()
```

```python
fig = px.histogram(df, x='B', title = 'Histograma da Variável B',
                   nbins = 30) # Está função define a quantidade de grupos para o gráfico
fig.update_layout(
    xaxis = dict(title = 'Variável A'),
    yaxis = dict(title = 'Frequência'),
)
fig.show()
```
