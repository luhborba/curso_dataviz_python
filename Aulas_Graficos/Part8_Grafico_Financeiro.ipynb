## Etapa 8: Gráficos Financeiros

```python
# Importações
from pandas_datareader import data
import plotly.graph_objects as go
import pandas as pd
import yfinance as yf
from datetime import datetime
```

```python
yf.pdr_override() #coloca o datareader do pandas como opicional
```

```python
# Importanto dados via API
dados = data.get_data_yahoo('PETR3.SA')
```

```python
#Visualizando dados
dados.head()
```

```python
# Criando variável com data atual
atual = datetime.now().strftime('%Y-%m-%d')
```

```python
#Criando Gráfico
linha = go.Scatter(
    x = dados.index,
    y = dados['Close'],
    name = 'PETR3 Close Price',
    mode = 'lines',
    line = dict(color = 'grey')
)

layout = go.Layout(
    title = 'PETR3 Close Price',
    xaxis = dict(title = 'Tempo'),
    yaxis = dict(title = 'Close Price'),
    xaxis_range = ['2020-01-01', atual] #Definindo um Range Padrão de Visualização
)

fig = go.Figure(linha, layout)

#Adicionando Range Slide
fig.update_xaxes(rangeslider_visible = True)
fig.show()
```
