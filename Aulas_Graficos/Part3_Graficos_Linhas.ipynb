## Etapa 3: Gráficos de Linhas


```python
# Importando Bibliotecas
import plotly.express as px
import plotly.graph_objs as go
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

### Aula 1

```python
#Criando vetor com 50 números inteiros
eixo_y = np.random.randint(100,5000,50)
eixo_y
```

```python
#Criando Index de 50 datas
eixo_x = pd.date_range(start='2021-1-1',freq='D', periods=50)
eixo_x
```

```python
# Iniciando Criação do Gráfico Estático
plt.figure(figsize = (8,5)) #Definindo tamanho do plot
plt.plot(eixo_x, eixo_y) #Definindo dados do gráfico e tipo
plt.title('Meu Gráfico') #Definindo titulo
plt.show() #Mostrando Gráfico
```

```python
# Criando mesmo Gráfico com Plotly
fig = px.line(x = eixo_x, y = eixo_y, title='Minha Figura')
fig.show()
```

### Aula 2

```python
#Criando Variável com DataFrame do Próprio Plotly
df = px.data.gapminder()
type(df)
```

```python
#Visualizando primeiras linhas
df.head()
```

```python
#Verificando Continentes existentes
df['continent'].unique()
```

```python
#Verificando Países existentes
df.country.unique()
```

```python
# Criando novo DataFrame filtrando apenas os dados referentes ao Brasil
df_filtrado = df.query('country == "Brazil"')
df_filtrado.head()
```

```python
# Verificando dados pós o filtro
df_filtrado.country.unique()
```

```python
# Criando Gráfico com Ploty de Linhas, mostrando o Crescimento da expectativa de vida através dos anos
title = 'Expectativa de Vida no Brasil (1952 - 2007)'
fig = px.line(df_filtrado, x = 'year', y = 'lifeExp', title=title)
fig.show()
```

```python
# Personalizando Gráfico
fig.update_layout(xaxis = dict(title = 'Tempo'),
                  yaxis = dict(title = 'Expectativa de Vida'),
                  plot_bgcolor = 'white'
                  )


#update_layout - Função para personalizar o gráfico
#xaxis - Função que segundo exemplo foi criando um dicionário usando o 'dict', passando 'title' para definir o titulo
#yaxis - Foi realizado similar para o eixo y
#plot_bgcolor - Função para definir a cor de fundo do plot
```

### Aula 3

```python
# Criando DataFrame com dados do próprio Ploty já filtrado
df = px.data.gapminder().query('continent == "Americas"')
title = 'Expectativa de Vida no Continente Americano (1957-2007)'
df.head()
```

```python
# Verificando Continente pós filtragem
df.continent.unique()
```

```python
# Criando Gráfico para visualizar a expectativa do continente Americano usando o Plotly
fig = px.line(df, x = 'year', y='lifeExp',
              title = title, color = 'country',)
fig.show()
```

```python
# Adicionando novas configurações ao gráfico
fig = px.line(df, x = 'year', y='lifeExp',
              title = title, color = 'country',
              width = 800,
              height = 400)
#Utilizando o 'width' e 'heigth' para definir o tamanho do gráfico
fig.update_layout(
    xaxis = dict(title = 'Ano'),
    yaxis = dict(title = 'Expectativa de Vida'),
    plot_bgcolor = 'white'
)
fig.show()
```

```python
# Criando Vetor com Numpy
x = np.arange(10)
```

```python
# Criando dados 'linspace' criando dados espaçados igualmente 'random.normal' criando números aleárotiros com desvio padrão de 2.5
N = 50
x = np.linspace(0,1,N)
random_y = np.random.normal(30, 2.5, N)
```

```python
# Criando 'linha' com dados, 'mode' definindo em gráfico de linha, 'name' definindo nome da linha, 'line' definindo configuração de linha, no caso foi utilizado o 'color' para escolher a cor
linha = [go.Scatter(
    x = x,
    y = random_y,
    mode = 'lines',
    name = 'Linha 1',
    line = dict(color = 'Black')
)]
# Criando configurações de Layout
layout = go.Layout(
    title = 'Minha Figura',
    xaxis = dict(title = 'Eixo X'),
    yaxis = dict(title = 'Eixo Y'),
)
# Mostrando Figura
fig = go.Figure(data = linha, layout = layout)
fig.show()
```

### Aula 6

```python
# Criando conjunto de dados de exemplo
N = 50
x = np.linspace(0,1,N) #Criando vetorr com números espaçados igualmente entre 0,1 com tamanho 50
random_y1 = np.random.normal(loc=30, scale=2.5, size=N) # Criando números aléatorios de distribuição normal, com média 30 e desvio padrão 2.5
random_y2 = np.random.randint(low=0, high=50, size=N) # Criando números aléatorios inteiros entre 0 e 50
random_y3 = np.random.uniform(low=0, high=20, size=N) # Criando números de distribuição uniform entre 0 e 20
```

```python
#Criando um grafico de múltiplas linhas
linha1 = go.Scatter(
    x = x,
    y = random_y1,
    mode = 'markers', #Formato de pontinhos invés de linhas
    name = 'Linha 1',
    line = dict(color = 'red')
)
linha2 = go.Scatter(
    x = x,
    y = random_y2,
    mode = 'lines+markers', #Formato de linha com marcadores
    name = 'Linha 2',
    line = dict(color = 'black', width = 1)
)
linha3 = go.Scatter(
    x = x,
    y = random_y3,
    mode = 'lines',
    name = 'Linha 3'
)

data = [linha1, linha2, linha3]

layout = go.Layout(
    title = 'Minha Figura',
    xaxis = dict(title='Eixo X'),
    yaxis = dict(title='Eixo Y')
)

fig = go.Figure(data, layout)
fig.show()
```

### Aula 7 e 8

```python
# Criando conjunto de dados utilizando funções numpy
N = 50
x = np.linspace(0,1,N) #Criando vetorr com números espaçados igualmente entre 0,1 com tamanho 50
random_y1 = np.random.normal(loc=30, scale=2.5, size=N) # Criando números aléatorios de distribuição normal, com média 30 e desvio padrão 2.5
random_y2 = np.random.randint(low=0, high=50, size=N) # Criando números aléatorios inteiros entre 0 e 50
random_y3 = np.random.uniform(low=0, high=20, size=N) # Criando números de distribuição uniform entre 0 e 20
```

```python
#Consultando site para verificação de cores
# www.rapidtables.com/web/color/gray-color.html
```

```python
# Criando gráfico ajustando diversas personalizações
fig = go.Figure()

fig.add_trace(go.Scatter(x = x,
                         y= random_y1,
                         mode = 'lines',
                         line = dict(color = '#D3D3D3', width = 2),
                         name = 'Linha 1'
                         ))
fig.add_trace(go.Scatter(x = x,
                         y = random_y2,
                         mode = 'lines+markers',
                         line = dict(color = '#2F4F4F', width = 2),
                         name = 'Linha 2',
                         marker = dict(size = 7, color = 'blue') #Alterando marker
                         ))
fig.add_trace(go.Scatter(x = x,
                         y = random_y3,
                         mode = 'lines',
                         line = dict(color = 'red', width = 2, dash = 'dot'), # utilizando dash para ajustar o tracejado da linha
                         name = 'Linha 3'
                         ))
fig.update_layout(plot_bgcolor = 'white',
                  title = 'Meu gráfico de Linhas',
                  xaxis = dict(title='Eixo X'),
                  yaxis = dict(title='Exito Y'),
                  font_family = 'Arial', #Definindo Fonte
                  font_color = 'black', # Definindo Cor da font
                  title_font_color = 'grey', #Definindo cor do título
                  legend_title = 'Linhas', #Criando título para legenda
                  legend_title_font_color = 'red' #Ajustando cor do título da legenda
                  )

fig.show()
```
