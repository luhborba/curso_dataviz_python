```python
# Criando DataFrame através de arquivo csv
df = pd.read_csv('train.csv')
```

```python
# Verificando as 5 primeiras Linhas
df.head()
```

```python
# Verificando o sumario estatístico
df.describe()
```

```python
# Realizando contagem de valores do campo Sex (Sexo)
df.Sex.value_counts()
```

```python
# Realizando Contagem de Sobreviventes 0 - Não Sobreviveu 1 - Sobreviveu
df.Survived.value_counts()
```

```python
# Verificando tamanho do DataFrame
df.shape
```

```python
# Filtrando apenas dados do sexo masculino
df[df['Sex']=='male']
```

```python
# Verificando eficacia do filtro anterior
df[df['Sex']=='male']['Sex'].unique()
```

```python
# Boa pratica
filtro_df = df[df['Sex']=='male']
filtro_df.shape
```

```python
# Outra forma de filtrar
filtro_df2 = df.loc[df['Sex']=='male']
filtro_df2.shape
```

```python
# 3 forma diferente de Fazer filtragem
filtro_df3 = df.query('Sex =="male"')
filtro_df3.shape
```

```python
# Multiplos Filtros
filtro_1 = df[(df['Sex']=='female') & (df['Survived']==1)]
filtro_1.head()
```

```python
#Verificando tamanho do conjunto
filtro_1.shape
```

```python
# Fazendo através do metódo query (& = And | = Or)
filtro_2 = df.query('(Sex == "female") & (Survived == 1)')
filtro_2.head()
```
